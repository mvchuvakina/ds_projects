```python
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']
```

<font color='green'> Отлично, что разделяешь категории на порядковые и номинальные.</font>

```python
#пайплайн для OneHotEncoder
ohe_pipe = Pipeline([
    (
        'ohe',
        OneHotEncoder(
            drop='first',
            sparse_output=False,
            handle_unknown='ignore'
        )
    )
])
```

```python
#пайплайн для OrdinalEncoder
ord_pipe = Pipeline([
    (
        'ord',
        OrdinalEncoder(
            categories=[
                ['junior', 'middle', 'sinior'],
                ['low', 'medium', 'high']
            ],  
            handle_unknown='use_encoded_value', 
            unknown_value=np.nan
        )
    ),
    (
        'simple_Imputer',
        SimpleImputer(
            missing_values=np.nan, 
            strategy='most_frequent'
        )
    )
])
```

<font color='blue'><b>Комментарий ревьюера : </b></font> ✔️\

```python
#единый пайплайн по подготовке данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)
```

## Обучение моделей

Нам нужно обучить 2 модели, из анализа матрицы корреляции нам известно, что есть признаки с высокой корреляцией (salary - level, workload, supervisor_evaluation - job_satisfaction_rate); мультиколлинеарности нет; целевая переменная — не бинарная, а непрерывная (job_satisfaction_rate); также наблюдаются выбросы в столбцах supervision_evaluation, salary. Исходя из этого в качестве линейной модели возьмем Ridge — она даёт стабильные результаты при умеренной корреляции между признаками. 

Для подбора гиперпараметров будем использовать RandomizedSearchCV, тк этот способ позволит подлбрать гиперпараметры быстрее, чем GridSearchCV, ведь у нас достаточно много гиперпараметров

```python
def smape(y_true, y_pred):
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    diff = np.abs(y_true - y_pred) / denominator
    return np.mean(diff) * 100



smape_scorer = make_scorer(smape, greater_is_better=False)

# Итоговый пайплайн
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', Ridge())  # базовая модель, заменится в RandomizedSearchCV
])

# Сетка параметров для моделей
param_grid = [
    {
        'models': [Ridge()],
        'models__alpha': [0.1, 1.0, 10.0],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [DecisionTreeRegressor()],
        'models__max_depth': range(5, 21, 2),
        'models__min_samples_split': [2, 5, 10, 20],
        'models__min_samples_leaf': [1, 2, 4],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

# RandomizedSearchCV
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_distributions=param_grid,
    cv=5,
    scoring=smape_scorer,
    n_iter=10,
    random_state=RANDOM_STATE,
    n_jobs=-1
)

# Обучение
randomized_search.fit(X_train, y_train)

# Лучшая модель
print('Лучшая модель и её параметры:\n')
print(randomized_search.best_estimator_)
print('\nSMAPE на тренировочной выборке (кросс-валидация):', -randomized_search.best_score_)
best_model = randomized_search.best_estimator_

# Прогноз на тестовой выборке
y_test_pred = best_model.predict(X_test)
final_smape = smape(y_test, y_test_pred)
print(f'\nSMAPE на тестовой выборке: {final_smape:.2f}')
```

```python
#пайплайн для OrdinalEncoder
ord_pipe = Pipeline([
    (
        'ord',
        OrdinalEncoder(
            categories=[
                ['junior', 'middle', 'sinior'],
                ['low', 'medium', 'high']
            ],  
            handle_unknown='use_encoded_value', 
            unknown_value=np.nan
        )
    ),
    (
        'simple_Imputer',
        SimpleImputer(
            missing_values=np.nan, 
            strategy='most_frequent'
        )
    )
])
```

```python
#единый пайплайн по подготовке данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)
```

## Обучение модели

Для подбора гиперпараметров будем использовать RandomizedSearchCV, тк этот способ позволит подлбрать гиперпараметры быстрее, чем GridSearchCV, ведь у нас достаточно много гиперпараметров

В данной задаче прогнозируется увольнение сотрудников ('yes': 1, 'no': 0), что делает задачу бинарной классификацией. При выборе метрики для оценки качества модели важно учитывать:

Несбалансированность классов: У нас наблюдается несбалансированность целевого признака, поэтому метрика точности не подойдет

ROC-AUC учитывает ранжирование вероятностей и не зависит от порога, что делает её устойчивой к дисбалансу классов.
позволяет корректно сравнивать различные модели.

Для обучения будем использовать следующие модели: DecisionTreeClassifier, KNeighborsClassifier, LogisticRegression также добавим автоматический отбор признаков SelectKBest, чтобы улучшить ROC-AUC.

```python
# Итоговый пайплайн: препроцессинг, отбор признаков и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('selector', SelectKBest(score_func=f_classif)),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

# Общий словарь параметров для RandomizedSearchCV
param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': [3, 5, 7],
        'models__max_features': [2, 5],
        'selector__k': [3, 5, 7, 9, 11, 'all'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': [3, 5, 7],
        'selector__k': [3, 5, 7, 9, 11, 'all'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE,
            solver='liblinear',
            penalty='l1')],
        'models__C': [0.1, 1, 10],
        'selector__k': [3, 5, 7, 9, 11, 'all'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

# Поиск по сетке
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    n_iter=20,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    verbose=2
)

# Обучение
randomized_search.fit(X_train, y_train)

# Лучшая модель и её параметры
print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print('Метрика ROC-AUC на кросс-валидации:', randomized_search.best_score_)

# Оценка на тестовой выборке
y_test_proba = randomized_search.predict_proba(X_test)[:, 1]
test_roc_auc = roc_auc_score(y_test, y_test_proba)
print(f'Метрика ROC-AUC на тестовой выборке: {test_roc_auc:.3f}')
```
