```python
#–û–∫—Ä—É–≥–ª—è–µ–º –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø–æ–µ–∑–¥–∫–∏
# common_data['rounded_duration'] = np.ceil(common_data['duration']).astype(int)
# common_data
```

```python
#5.1 –°–æ–∑–¥–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ 
monthly_data = common_data.groupby(['user_id', 'month']).agg(
    total_distance=('distance', 'sum'),  
    total_rides=('duration', 'count'),  
    total_duration=('duration', 'sum'),    
    start_ride_price=('start_ride_price', 'first'),  
    minute_price=('minute_price', 'first'),
    subscription_fee=('subscription_fee', 'first'),
#    rounded_duration=('rounded_duration', 'sum')  
).reset_index()
monthly_data 
```

```python
#–°–æ–∑–¥–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é, —á—Ç–æ–±—ã –ø–æ—Å—á–∏—Ç–∞—Ç—å –µ–∂–µ–º–µ—Å—è—á–Ω—É—é –≤—ã—Ä—É—á–∫—É
def calculate_revenue(row):
    revenue = (row['start_ride_price'] * row['total_rides'] + 
               row['minute_price'] * row['total_duration'] +
               row['subscription_fee'])
    return revenue

# –ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –≤—ã—Ä—É—á–∫–∏
monthly_data['revenue'] = monthly_data.apply(calculate_revenue, axis=1)
monthly_data
```
